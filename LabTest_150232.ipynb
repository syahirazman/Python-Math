```python
#Q3
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(2*-np.pi, 2*np.pi, 0.0001)
y1 = np.cos(x)
y2 = np.sin(0.5*x)
y3 = np.cos(3*x)

plt.plot(x,y1, label = "y = cos t", color = 'blue')
plt.plot(x,y2, label = "y = sin 1/2 t", color = 'orange')
plt.plot(x,y3, label = "y = cos 3t", color = 'green')

plt.xlabel('t')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
#Correction Q3
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(2*-np.pi, 2*np.pi, 0.1)
y1 = np.cos(x)
y2 = np.sin(0.5*x)
y3 = np.cos(3*x)

plt.plot(x,y1, label = "y = cos t", color = 'blue')
plt.plot(x,y2, label = "y = sin 1/2 t", color = 'orange')
plt.plot(x,y3, label = "y = cos 3t", color = 'green')

plt.xlim([-2*np.pi, 2*np.pi]) 
plt.ylim([-1, 1])

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
          [r'$-2\pi$', r'$-1\pi$', r'$0$', r'$1\pi$', r'2$\pi$'])
plt.yticks([-1, 0, 1])

plt.xlabel('t')
plt.ylabel('y')
plt.grid()

plt.legend()
plt.show()
```
